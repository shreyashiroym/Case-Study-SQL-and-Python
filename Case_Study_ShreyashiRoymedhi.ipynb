### Question 5

### Question 6

```python
pd.read_sql_query('SELECT DISTINCT COUNT("first_name"||"last_name") AS "count", "state", SUM(" cash_fine ") AS "cash_total" FROM fines GROUP BY "state" ORDER BY "count" DESC LIMIT 1', con=dbconn)
```

### Question 7
